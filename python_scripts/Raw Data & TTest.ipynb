## TABLE 2
Below is to check the p-values between values in Table 2 to check the numerical significance between them and if they're insignificantly different. 

```python
from scipy.stats import ttest_rel
print("SVM ACC & LOGREG ACC:")
print(ttest_rel(df_w_index.iloc[3],df_w_index.iloc[0]),"\n")
print("SVM ACC & DT ACC:")
print(ttest_rel(df_w_index.iloc[6],df_w_index.iloc[0]),"\n")
```

```python
print("SVM ROC AUC & LOGREG ROC AUC:")
print(ttest_rel(df_w_index.iloc[4],df_w_index.iloc[1]),"\n")
print("SVM ROC AUC & DT ROC AUC:")
print(ttest_rel(df_w_index.iloc[7],df_w_index.iloc[1]),"\n")
```
